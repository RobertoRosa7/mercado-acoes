<h1>Como pegar cotação de ações com python</h1>

```python
from pandas_datareader import data as web
import pandas as pd
import matplotlib.pyplot as plt
```

```python
cotacao_bovespa = web.DataReader('^BVSP', data_source='yahoo', start="01-01-2019", end="01-01-2022")
cotacao_bovespa["Company"] = "BVSP"
cotacao_bovespa.reset_index()
# cotacao_bovespa.to_csv('../data/files/cotacoes.csv')
```

```python
companies = pd.read_csv(
    '../data/files/empresas_cotacoes.csv', sep=',', encoding="UTF-8")
df = pd.DataFrame()

for company in companies['name_company']:
    try:
        cotacao = web.DataReader(
            f'{company}.SA', data_source='yahoo', start="01-01-2019", end="01-01-2022")
        cotacao["Company"] = company

        df = pd.concat([df, cotacao])
    except Exception as e:
        print(f'Error na busca de {company}\n {e}')
```

```python
df.reset_index()
```

```python
df.to_csv('../data/files/cotacoes.csv')
```
